```python

import ast

import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on='title')

```

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies.dropna(inplace=True)
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

def fetch_director(obj):              # ← you also had this missing!
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            return [i['name']]
    return []

def fetch_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L

movies['genres']   = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['cast']     = movies['cast'].apply(fetch_cast)
movies['crew']     = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
  # Convert overview sentence into words



# Remove spaces so multi-word terms stay together
for col in ['genres', 'keywords', 'cast', 'crew']:
    movies[col] = movies[col].apply(lambda x: [i.replace(" ", "") for i in x])


#Combine all columns into one tags column
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']

# Create a clean new DataFrame
new_df = movies[['movie_id', 'title', 'tags']].copy()
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
#nltk = Natural Language Toolkit — a Python library for text processing
#PorterStemmer = a specific stemming algorithm (invented by Martin Porter in 1980). It uses rules like "remove -ing, -ed, -es" to chop word endin
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

new_df['tags'] = new_df['tags'].apply(stem)




#What is Stemming?
#Before the code — understand the core idea:
#####Stemming cuts words down to their root form.
##"loved"   →  "love"
##"dancing" →  "danc"
##"action"  →  "action"
##"actions" →  "action"
##Why does this help? Without stemming, your model treats "love", "loved", and "loving" as 3 different words. With stemming they all become "love" — so movies with any of these words match each other correctly.
```

```python
# vectorize 
#Vectorization means converting data (especially text, images, categorical values, etc.) into numerical vectors
#(arrays of numbers) so that algorithms can understand and process them



from sklearn.feature_extraction.text import CountVectorizer
  #convert text into numbers
   #create vectors from words
from sklearn.metrics.pairwise import cosine_similarity
   # used to measure similarity between two movies 
cv = CountVectorizer(max_features=5000, stop_words='english')
   #it creates vector
vectors = cv.fit_transform(new_df['tags']).toarray()
similarity = cosine_similarity(vectors)
print(similarity.shape)
```

```python
## max_features=5000 -> keep top 5000 important words only
# stop_words='english' -> remove common English words like "the", "is", "and"
## fit_transform() learns vocabulary and transforms text into vectors
# toarray() converts sparse matrix into normal numpy array
# Calculate cosine similarity between all movie vectors
# Higher value means movies are more similar

## Print shape of similarity matrix
# Example: (4806, 4806) means 4806 movies compared with all 4806 movies
```

```python
def recommend(movie):
     # function to reccomend movies to the given movie nakme
    movie_index = new_df[new_df['title'] == movie].index[0]
     # find the index of the movie by the user 
    distances = similarity[movie_index]
      #gets the  similarity score of the selected movie 
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
       #sort movies based on similarity score 
       #loop through the reccomnded movies 
    for i in movies_list:
        print(new_df.iloc[i[0]].title)

# Test it!
recommend('Baahubali: The Beginning')
```
